```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import MDS
from sklearn.metrics.pairwise import rbf_kernel

# 임의의 2차원 데이터 생성
X = np.random.rand(10, 2)

# MDS를 사용하여 데이터를 1차원으로 임베딩
embedding = MDS(n_components=1)
X_embedded = embedding.fit_transform(X)

# 커널 추정 (RBF 커널)
kernel_matrix = rbf_kernel(X)

# 결과 시각화
plt.scatter(X_embedded, np.zeros_like(X_embedded), c='b', label='Embedded Data')
plt.legend()
plt.title('Embedded Data with RBF Kernel')
plt.show()
```
