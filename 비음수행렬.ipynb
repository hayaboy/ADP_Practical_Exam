```python
from sklearn.decomposition import NMF
import numpy as np

# 예제 데이터 생성
V = np.array([[1, 2, 3],
              [4, 5, 6]])

# NMF 모델 생성 및 학습
model = NMF(n_components=2, init='random', random_state=0)
W = model.fit_transform(V)
H = model.components_

print("Original matrix V:")
print(V)
print("\nMatrix W (components):")
print(W)
print("\nMatrix H (basis vectors):")
print(H)

# 재구성된 행렬 확인
reconstructed_V = np.dot(W, H)
print("\nReconstructed matrix V (W dot H):")
print(reconstructed_V)
```
