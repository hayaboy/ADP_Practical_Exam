```python
import numpy as np
import matplotlib.pyplot as plt

# x 값 생성
x = np.linspace(-10, 10, 400)  # -10부터 10까지 400개의 점을 생성

# f(x) 정의
def f(x):
    return 3 * x + 1

# y 값 계산
y = f(x)

# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = 3x + 1', color='blue')

# 그래프에 제목, 축 레이블, 범례 추가
plt.title('Graph of f(x) = 3x + 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# 그래프 보여주기
plt.show()
```

```python
#numpy의 log는 자연로그이므로 밑수가 1/2인 것을 표현하려면
#  np.log(x)/ np.log(0.5) 해야함
```

```python

import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-10,10)
y=np.log(x) / np.log(0.5)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$\log_{\frac{1}{2}} x$', color='blue')
plt.title(r'Graph of $\log_{\frac{1}{2}} x$')
plt.xlabel('x')
plt.ylabel(r'$\log_{\frac{1}{2}} x$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import warnings
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')

x=np.linspace(-10,10)
y=np.log(x) / np.log(0.5)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$\log_{\frac{1}{2}} x$', color='blue')
plt.title(r'Graph of $\log_{\frac{1}{2}} x$')
plt.xlabel('x')
plt.ylabel(r'$\log_{\frac{1}{2}} x$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
