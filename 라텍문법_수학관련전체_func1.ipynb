```python
import numpy as np
import matplotlib.pyplot as plt

# x  
x = np.linspace(-10, 10, 400)  # -10 10 400  

# f(x) 
def f(x):
    return 3 * x + 1

# y  
y = f(x)

#  
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = 3x + 1', color='blue')

#  ,  ,  
plt.title('Graph of f(x) = 3x + 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

#  
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-10,10)
y=np.log(x) / np.log(0.5)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$\log_{\frac{1}{2}} x$', color='blue')
plt.title(r'Graph of $\log_{\frac{1}{2}} x$')
plt.xlabel('x')
plt.ylabel(r'$\log_{\frac{1}{2}} x$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import warnings
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')

x=np.linspace(-10,10)
y=np.log(x) / np.log(0.5)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$\log_{\frac{1}{2}} x$', color='blue')
plt.title(r'Graph of $\log_{\frac{1}{2}} x$')
plt.xlabel('x')
plt.ylabel(r'$\log_{\frac{1}{2}} x$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
