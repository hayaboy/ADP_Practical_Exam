```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
dat=sns.load_dataset('mpg')
```

```python
dat.shape
```

```python
dat.head()
```

```python
sns.distplot(dat.mpg, kde=False, rug=False, bins=50, color='red')
plt.show()
```

커널 밀도 추정(Kernel Density Estimation, KDE)은 주어진 데이터셋의 확률 밀도 함수(probability density function, PDF)를 추정하는 비모수적(non-parametric) 방법 중 하나입니다.

간단히 말해, 주어진 데이터셋에서 각 데이터 포인트를 중심으로 일종의 커널(예를 들면 가우시안 커널)을 그려서, 이 커널들의 합으로 전체 데이터의 분포를 근사합니다. 이렇게 하면 데이터의 분포를 부드럽게 추정할 수 있습니다.

KDE는 데이터를 시각화할 때 유용하게 활용됩니다. 특히 히스토그램과 함께 사용되면 데이터의 분포를 더 부드럽게 표현하여 데이터셋의 특징을 더 잘 이해할 수 있습니다. 그러나 KDE는 대량의 데이터에 대해 연산 비용이 많이 소요될 수 있으므로 데이터가 매우 큰 경우에는 사용에 주의해야 합니다.

```python
sns.distplot(dat.mpg, kde=True, rug=False, bins=50, color='red')
plt.show()
```

```python
sns.distplot(dat.mpg, kde=True, rug=True, bins=50, color='red')
plt.show()
```

```python
sns.kdeplot(dat.mpg, color='blue')
plt.show()
```

```python
sns.jointplot(x='weight', y='mpg', data=dat, color='red',kind='scatter')
plt.show()
```

```python
sns.jointplot(x='weight', y='mpg', data=dat, color='orange',kind='reg')
plt.show()
```

```python
sns.lmplot(data=dat, x='weight', y='mpg')
plt.show()
```

```python
sns.lmplot(data=dat, x='weight', y='mpg', hue='origin')
plt.show()
```

```python
sns.lmplot(data=dat, x='weight', y='mpg', col='origin', hue='origin')
plt.show()
```

```python
# help(sns.lmplot)
```

```python
sns.lmplot(data=dat, x='weight', y='mpg', col='origin', hue='origin',\
           markers=['o','v','^'], scatter_kws={'s':30}, aspect=0.5)
plt.show()
```

size 매개변수는 seaborn 버전 0.11.0 이후에 scatter_kws를 사용하여 설정할 수 있습니다. 이전 버전에서는 size 매개변수를 사용하여 점의 크기를 조절했지만, 현재 버전에서는 scatter_kws를 사용하여 더 일반적인 설정을 수행합니다.

따라서 scatter_kws={'s':30}는 각 점의 크기를 지정하는 것입니다. 여기서 's'는 점의 크기를 나타내며, 30은 점의 크기를 의미합니다

서브플롯의 종횡비는 서브플롯의 가로 길이와 세로 길이 사이의 비율을 나타냅니다. 예를 들어, aspect=0.5는 서브플롯의 가로 길이가 세로 길이의 절반인 정사각형을 나타내며, aspect=2는 세로 길이가 가로 길이의 두 배인 서브플롯을 나타냅니다.

```python
sns.jointplot(x='weight', y='mpg', data=dat, color='green', kind='hex')
plt.show()
```

```python
sns.jointplot(x='weight', y='mpg', data=dat, color='blue', kind='kde')
plt.show()
```

```python
sns.barplot(x='origin', y='mpg', data=dat)
plt.show()
```

막대 그래프의 가운데 까만 작은 선은 해당 막대의 95% 신뢰 구간(confidence interval)을 나타냅니다. 이 신뢰 구간은 해당 그룹의 평균에 대한 추정치의 신뢰도를 시각적으로 표현합니다. 일반적으로 이 신뢰 구간은 데이터의 변동성을 고려하여 평균 추정치 주변의 신뢰 구간을 나타내며, 오차 막대(error bars)로도 알려져 있습니다.

이 신뢰 구간을 제거하는 것은 반드시 필요한 것은 아닙니다. 그러나 때로는 시각적으로 혼란스럽게 만들 수 있기 때문에 이를 제거하는 것이 유용할 수 있습니다.

가운데 선을 제거하려면 ci=None 매개변수를 barplot 함수에 전달하면 됩니다. 따라서 코드는 다음과 같이 수정될 수 있습니다:

```python
sns.barplot(x='origin', y='mpg', data=dat, ci=None)
plt.show()
```
