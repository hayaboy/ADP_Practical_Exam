```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs

# 데이터 생성
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# GMM 적용
gmm = GaussianMixture(n_components=4, random_state=0)
gmm.fit(X)
labels = gmm.predict(X)

# 클러스터링 결과 시각화
plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis')
plt.title('GMM Clustering')
plt.show()

# 각 가우시안 분포의 평균과 공분산
print("Means:", gmm.means_)
print("Covariances:", gmm.covariances_)
```

이 예시에서는 make_blobs 함수를 사용하여 4개의 클러스터를 가지는 데이터를 생성하고, GMM을 사용하여 클러스터링을 수행합니다. 각 데이터 포인트가 할당된 클러스터를 시각화하고, 각 가우시안 분포의 평균과 공분산을 출력합니다.

결론
Gaussian Mixture Model (GMM)은 데이터를 여러 가우시안 분포의 혼합으로 모델링하는 강력한 클러스터링 기법입니다. EM 알고리즘을 사용하여 파라미터를 추정하고, 각 데이터 포인트가 각 클러스터에 속할 확률을 계산합니다. GMM은 데이터의 복잡한 분포를 효과적으로 모델링할 수 있지만, 초기 설정에 민감하고 계산 비용이 많이 들 수 있습니다. 데이터의 특성과 분석 목적에 따라 GMM을 적절히 사용하면 유용한 도구가 될 수 있습니다.
