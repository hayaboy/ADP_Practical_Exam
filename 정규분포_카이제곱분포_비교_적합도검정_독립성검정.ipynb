```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

#    
n_samples = 10000
k = 3  #  (    )

#   k    
chi_square_samples = np.sum(np.random.normal(size=(n_samples, k))**2, axis=1)

#   
plt.hist(chi_square_samples, bins=50, density=True, alpha=0.6, color='g', label='Empirical Chi-Square')

#    
x = np.linspace(0, 20, 1000)
plt.plot(x, stats.chi2.pdf(x, df=k), 'r-', label='Theoretical Chi-Square (df={})'.format(k))

#  
mean = np.mean(chi_square_samples)
std_dev = np.std(chi_square_samples)
x_norm = np.linspace(0, 20, 1000)
plt.plot(x_norm, stats.norm.pdf(x_norm, mean, std_dev), 'b--', label='Normal Distribution')

plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.title('Chi-Square Distribution with {} Degrees of Freedom and Normal Distribution'.format(k))
plt.show()
```

```python
import numpy as np
from scipy.stats import chisquare

#  
observed = np.array([68, 32])
expected = np.array([50, 50])

#   
chi2, p = chisquare(f_obs=observed, f_exp=expected)

#  
print(f"Chi2 Statistic: {chi2}")
print(f"P-value: {p}")
```

```python
import scipy.stats as stats

#   
test_statistic = 143.407
df = 1

#  0.01 (99% )
alpha = 0.01

#  
p_value = 1 - stats.chi2.cdf(test_statistic, df)
critical_value = stats.chi2.ppf(1 - alpha, df)

#  
print("P-value:", p_value)
print("Critical Value:", critical_value)

#    
if p_value < alpha:
    print(":  .")
else:
    print(":    .")
```
