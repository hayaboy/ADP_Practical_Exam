```python
import os
import pandas as pd
from googleapiclient.discovery import build

# API 키 설정
api_key = ''

# YouTube API 클라이언트 생성
youtube = build('youtube', 'v3', developerKey=api_key)

def get_comments(video_id):
    comments = []
    # 첫 번째 API 요청 - 댓글 리스트
    response = youtube.commentThreads().list(
        part='snippet',
        videoId=video_id,
        maxResults=5,  # 한 번에 가져올 댓글 수 (최대 100)
        textFormat='plainText'
    ).execute()

    while response:
        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comments.append([comment['authorDisplayName'], comment['textDisplay']])

        # 다음 페이지가 있는 경우 다음 페이지로 이동
        if 'nextPageToken' in response:
            response = youtube.commentThreads().list(
                part='snippet',
                videoId=video_id,
                pageToken=response['nextPageToken'],
                maxResults=5,
                textFormat='plainText'
            ).execute()
        else:
            break

    return comments

# 테스트: 특정 비디오 ID의 댓글 가져오기
video_id = ''  # 비디오 ID를 여기에 입력하세요
comments = get_comments(video_id)

# 데이터프레임으로 변환
df = pd.DataFrame(comments, columns=['Author', 'Comment'])

# CSV 파일로 저장
df.to_csv('youtube_comments.csv', index=False, encoding='utf-8')

print('댓글을 성공적으로 가져와서 youtube_comments.csv 파일로 저장했습니다.')
```
